```python
# Handle Missing Keys with ## setdefault() and defaultdict()
# setdefault() 
periodic_table = {'Hydrogen': 1, 'Helium': 2}
print(periodic_table)
carbon = periodic_table.setdefault('Carbon', 12)  # setdefault -> add new key and the new value is used
print(carbon)
print(periodic_table)                             # {'Hydrogen': 1, 'Helium': 2, 'Carbon': 12}

helium = periodic_table.setdefault('Helium', 947)
print(helium)       # 對 existing key : the original value is returned and nothing is changed
```

```python
from collections import defaultdict    # defaultdict() is similar to  setdefault()

periodic_table = defaultdict(int)          # defaultdict 
periodic_table['Hydrogen'] =1              # add key & value
periodic_table['Lead']                     # 不給value ，Value = 0
print(periodic_table)
```

```python
from collections import defaultdict

def no_idea():
   return 'Huh?'

bestiary = defaultdict(no_idea)
bestiary['A'] = 'Abominable Snowman'
bestiary['B'] = 'Basilisk'
print(bestiary['A'])
print(bestiary['B'])

bestiary['B']

bestiary['C']

print(bestiary)

bestiary = defaultdict(lambda: 'Huh?')
bestiary['E']

print(bestiary)
```

```python
from collections import defaultdict
food_counter = defaultdict(int)
for food in ['spam', 'spam', 'eggs', 'spam']:
    food_counter[food] += 1

for food, count in food_counter.items():
    print(food, count)
```

```python
# 一般的 dictionary 不是 defaultdict 寫法 : 不然會有錯誤
dict_counter = {}
for food in ['spam', 'spam', 'eggs', 'spam']:
    if not food in dict_counter:
        dict_counter[food] = 0
    dict_counter[food] += 1
    for food, count in dict_counter.items():
        print(food, count)
```

```python
# Count Items with Counter()    return dictionary
from collections import Counter
breakfast = ['spam', 'spam', 'eggs', 'spam']
breakfast_counter = Counter(breakfast)
print(breakfast_counter)
breakfast_counter['spam']
```

```python
# The most_common() function returns all elements in descending order
print(breakfast_counter.most_common())

print(breakfast_counter.most_common(1))
```

```python
lunch = ['eggs', 'eggs', 'bacon']
lunch_counter = Counter(lunch)
print(lunch_counter)
print(breakfast_counter + lunch_counter)    # breakfast and lunch 有甚麼
print( breakfast_counter - lunch_counter)   # breakfast 有 but lunch 沒有
print(lunch_counter - breakfast_counter)
print(breakfast_counter & lunch_counter)
breakfast_counter | lunch_counter
```

```python
# Stack + Queue == deque  :
# The function popleft() removes the leftmost item from the deque and returns it;
# pop() removes the rightmost item and returns it.  
def palindrome(word):
    from collections import deque
    dq = deque(word)
    while len(dq) > 1:
        if dq.popleft() != dq.pop():
            return False
    return True

print(palindrome('a'))

print(palindrome('racecar'))

print(palindrome(''))

print(palindrome('radar'))

print(palindrome('halibut'))

def another_palindrome(word):
    return word == word[::-1]
...
>>> another_palindrome('radar')
True
>>> another_palindrome('halibut')
```

```python
# Iterate over Code Structures with itertools : 用itertools 迭代程式結構
import itertools

for item in itertools.chain([1, 2], ['a', 'b'],['r','f','d','f','g','g']):
    print(item)
```

```python
# cycle() is an infinite iterator, cycling through its arguments:
for item in itertools.cycle([1, 2]):   
    print(item)
```

```python
# accumulate() calculates accumulated values. By default, it calculates the sum: (iterate sum)
import itertools
for item in itertools.accumulate([1, 2, 3, 4]):
    print(item)

print("========================================")
def multiply(a, b):
    return a * b

for item in itertools.accumulate([1, 2, 3, 4], multiply):
    print(item)
```

```python
# Get Random
from random import choice
print(choice([23, 9, 46, 'bacon', 0x123abc]))

print(choice( ('a', 'one', 'and-a', 'two') ))

print(choice(range(100)))

print(choice('alphabet'))

print("====================================")
# Use the sample() function to get more than one value at a time:
from random import sample
print(sample([23, 9, 46, 'bacon', 0x123abc], 3))

print(sample(('a', 'one', 'and-a', 'two'), 2))

print(sample(range(100), 4))

print(sample('alphabet', 7))

print("====================================")

# To get a random integer from any range, you can use choice() or sample() with range(), or use randint() or randrange():
from random import randint
print(randint(38, 74))
print(randint(38, 74))
print(randint(38, 74))

from random import randrange
print(randrange(38, 74))
print(randrange(38, 74, 10))

print(randrange(38, 74, 10))
print("====================================")

# Finally, get a random real number (a float) between 0.0 and 1.0:
from random import random
print(random())

print(random())

print(random())
```

```python
# step1 : Rectangle.py 內容如下
class Rectangle():       #定義父類別  
    def __init__(self, width,height):
        self.width = width    #定義共用屬性   
        self.height = height  #定義共用屬性
    def area(self):           #定義共用方法 
        return self.width * self.height
```

```python
# step2 : Triangle.py 內容如下
from areapackage2.Rectangle import Rectangle
        
class Triangle(Rectangle): #定義子類別  
    def area2(self):       #定義子類別的共用方法 
        return (self.width * self.height)/2
```
