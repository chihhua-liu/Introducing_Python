## Open
Create or Open with open()
1. Read an existing file
2. Write to a new file
3. Append to an existing file
4. Overwrite an existing file

fileobj = open( filename, mode )
1. fileobj is the file object returned by open()
2. filename is the string name of the file
3. mode is a string indicating the file’s type and what you want to docwith it

1. The first letter of mode indicates the operation:
     1. r means read.
     2. w means write. If the file doesn’t exist, it’s created. If the file does exist, it’s overwritten.
     3. x means write, but only if the file does not already exist.
     4. a means append (write after the end) if the file exists.

2. The second letter of mode is the file’s type: 
    1. t (or nothing) means text.  :default
    2. b means binary

```python
# seek(num) , tell()
f=open('file.bin','rb')
print("目前文件索引位置：",f.tell()) #0
f.seek(6) #移到索引第 6 (第7個字元)位置
str1=f.read(7) #讀取 7 個字元
print(str1)  # b'Python\n'
print("目前文件索引位置：",f.tell()) #13

f.seek(0) #回文件最前端
print("目前文件索引位置：",f.tell()) #0
str2=f.read(5) #讀取 5 個字元
print(str2)  # b'Hello'

f.seek(-8,2) #移至最尾端，向前取 8 個字元
str3=f.read()
print(str3)  # b'Welcome\n'

f.close()
```

```python
# To make print
# work like write, pass it the following two arguments:
#  1. sep (separator, which defaults to a space, ' ')
#  2. end (end string, which defaults to a newline, '\n')
fout = open('relativity', 'wt')
print(poem, file=fout, sep=", end=")
fout.close()
```

```python
poem = '''There was a young lady named Bright,
Whose speed was far faster than light;
She started one day
In a relative way,
And returned on the previous night.'''
print(len(poem))
# If you have a large source string, you can also write chunks (using slices) until the source is done:
fout = open('relativity', 'wt')
size = len(poem)
offset = 0
chunk = 100
while True:
    if offset > size:
        break
    fout.write(poem[offset:offset+chunk])   # 一次寫100 字
    offset += chunk
fout.close()
```

```python
try:
    fout = open('relativity', 'xt')   # x if file exist : error ， so used try
    fout.write('stomp stomp stomp')
except FileExistsError:
    print('relativity already exists!. That was a close one.')
```

```python
# Read a Text File with read(), readline(), or readlines()
fin = open('relativity', 'rt' )
poem = fin.read()
fin.close()
len(poem)
print(poem)
```

```python
# chunk(大的數量) to a poem string to rebuild the original:
poem = ''
fin = open('relativity', 'rt' )
chunk = 100
while True:
    fragment = fin.read(chunk)
    if not fragment:
        break
    poem += fragment

fin.close()
len(poem)
print(poem)
```

```python
# You can also read the file a line at a time by using readline()
poem = ''
fin = open('relativity', 'rt' )
while True:
    line = fin.readline()
    if not line:
        break
    poem += line

fin.close()
len(poem)
print(poem)
```

```python
# The easiest way to read a text file is by using an iterator.
poem = ''
fin = open('relativity', 'rt')
for line in fin:
    poem +=line
fin.close()
print(poem)
```

```python
# readlines()
fin = open('relativity', 'rt' )
lines = fin.readlines()
fin.close()
print(len(lines), 'lines read')
print(lines)
print("==================================")
for line in lines:
    print(line, end='')
```

## Binary File

```python
# Write a Binary File with write()
bdata = bytes(range(0, 256))
print(len(bdata))
print(bdata)

fout = open('bfile', 'wb')  # b is binary
fout.write(bdata)
fout.close()
```

```python
fout = open('bfile', 'wb')
size = len(bdata)
offset = 0
chunk = 100
while True:
    if offset > size:
        break
    fout.write(bdata[offset:offset+chunk])
    offset += chunk
fout.close()
```

```python
# Read a Binary File with read()
fin = open('bfile', 'rb')
bdata = fin.read()
fin.close()
len(bdata)
print(bdata)
```

```python
# Close Files Automatically by Using with
# 1. Python has context managers to clean up things such as open files. You use the form with 
# 2. expression as variable:
with open('relativity', 'wt') as fout:
    fout.write(poem)
```

```python
# Change Position with seek()
# The tell() function returns your current offset from the beginning of the file,in bytes. 
# The seek() function lets you jump to another byte offset in the file.
fin = open('bfile', 'rb')
print(fin.tell())    # index[0]
print(fin.seek(245))
bdata = fin.read()
print(len(bdata))
bdata[10]   # index[0] from 245  
```

```python
# seek(offset, origin): offset 位移幾個 index
#  origin : 
# (1) If origin is 0 (the default), go offset bytes from the start
# (2) If origin is 1, go offset bytes from the current position
# (3) If origin is 2, go offset bytes relative to the end
fin = open('bfile', 'rb')
# print(fin.tell())
print(fin.seek(-250,2)) # 距離End point is 250
bdata = fin.read()
print(len(bdata))
print(bdata)
fin.close()
```
