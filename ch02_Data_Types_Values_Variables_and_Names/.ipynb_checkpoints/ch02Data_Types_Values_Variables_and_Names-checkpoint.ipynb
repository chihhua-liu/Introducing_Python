```python
# 初始化
string01 = "1,2,3,4"
print(string01)
type(string01)
```

```python
# 字串分割
'''
用法
string.split()
說明
默認以空格、換行字元分割字串s，返回 list
'''
list01 = string01.split(',')
print(list01)
```

```python
# 將 list 元素合併成字串
'''
用法
string.join(seq)
說明
以string為分隔符，將seq中的元素串起來成為一個新的字串
'''
string02 = '-'.join(list01)
print("將 list 元素合併成字串的結果: {}".format(string02))

string02 = ''.join(list01)
print("將 list 元素合併成字串的結果: {}".format(string02))
```

```python
# 搜尋字串
string03 = 'believe'
result03 = string03.find('lie')
'''
用法
string.find(str)
說明
返回s第一次在字串s中出現的index，若找不到則返回-1
'''
print(result03)
```

```python
# 替換字串
'''
用法
string.replace(str1, str2)
說明
將s中的str1替換成str2
'''
string04 = "Alex"
string04 = string04.replace('lex', 'llen')
print(string04)
```

```python
# 字串變成小寫
'''
用法
string.lower()
說明
將字串s裡的字母全部改成小寫
'''
print("CAR".lower())
```

```python
# 字串變成大寫
'''
用法
string.upper()
說明
將字串s裡的字母全部改成大寫
'''
print("good".upper())
```

```python
# 去除字串左邊空格
'''
用法
string.lstrip()
說明
去除字串s左邊的空格
'''
string05 = "         ___ccc___"
print("原先的字串: {}".format(string05))
print("去除左邊空格後: {}".format( string05.lstrip() ))
```

```python
# 去除字串右邊空格
'''
用法
string.rstrip()
說明
去除字串s右邊的空格
'''
string06 = "___ccc___         "
print("原先的字串: {}".format(string06))
print("去除左邊空格後: {}".format( string06.rstrip() ))
```

```python
# 去除字串兩側空格
'''
用法
string.strip()
說明
去除字串s左、右兩邊的空格
'''
string07= "        ___ccc___         "
print("原先的字串: {}".format(string07))
print("去除左邊空格後: {}".format( string07.strip() ))
```
