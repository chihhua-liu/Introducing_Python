# The datetime Module
It defines four main object classes, each with many methods:
(1) date for years, months, and days
(2) time for hours, minutes, seconds, and fractions
(3) datetime for dates and times together
(4) timedelta for date and/or time intervals

```python
from datetime import date
halloween = date(2019, 10, 31)
print(halloween)
print(halloween.day)
print(halloween.month)
print(halloween.year)

halloween.isoformat()      # isoformat()

now = date.today()         # date.today()
print(now)
now.isoformat()
```

```python
# This one makes use of a timedelta object to add some time interval to a date
from datetime import timedelta
one_day = timedelta(days=1)      # timedelta(days=5) 間隔幾天(5 days)，可以用在
tomorrow = now + one_day
print(tomorrow)

print(now + 17*one_day)

yesterday = now - one_day
print(yesterday)
```

```python
from datetime import time
noon = time(12, 15 , 30, 30)
print(noon)

print(noon.hour)

print(noon.minute)
print(noon.second) 
print(noon.microsecond)   #微秒，（英文：microsecond，符號µs），1微秒等如1000000分之1秒（一百萬分之一秒）

from datetime import datetime
dateTimeObj = datetime.now()
print(dateTimeObj)
print(dateTimeObj.second)
```

```python
from datetime import datetime
some_day = datetime(2019, 1, 2, 3, 4, 5, 6)
print(some_day)

# The datetime object also has an isoformat() method:
some_day.isoformat()
```

```python
now = datetime.now()
print('now = {}'.format(now))
print('now.year = {}'.format(now.year))
print('now.month = {}'.format(now.month))
print('now.day = {}'.format(now.day))
print('now.hour = {}'.format(now.hour))
print('now.minute = {}'.format(now.minute))
print('now.second = {}'.format(now.second))
print('now.microsecond = {}'.format(now.microsecond))

# You can combine() a date object and a time object into a datetime:
from datetime import datetime, time, date
noon = time(12)

this_day = date.today()
noon_today = datetime.combine(this_day, noon)
print('noon_today ={} '.format(noon_today))

print(noon_today.date())

print(noon_today.time())
```

# Using the time Module


```python
# Unix time uses the number of seconds since midnight on January 1, 1970. This value is often called the epoch,
import time
now = time.time()
print(now)
# You can convert an epoch value to a string by using ctime()
time.ctime(now)
```

```python
# My 15:19 (Central time zone, Daylight Savings)
import time
nowlocal = time.localtime(now) 
print(nowlocal)

# UTC (formerly called Greenwich time or Zulu time)
print(time.gmtime(now))


now = time.localtime()
print(now)
print(now[0])
print(list(now[x] for x in range(9)))

# mktime() goes in the other direction, converting a struct_time object to epoch seconds:
//tm = time.localtime(now)
//timetest = time.mktime(tm)
//print(timetest )
```

```python
import time
fmt = "It's %A, %B %d, %Y, local time %I:%M:%S%p" # first define the format string fmt  
t = time.localtime()
t
```
